# Applying Train Test Split on the Data

```python
# Spliiting data in x and y
def splitting_data(dataframe):
    global x, y
    x = dataframe.drop(columns = ['Probability'],axis =1)
    y = dataframe['Probability']
```

```python
from sklearn.model_selection import train_test_split
```

```python
def tts(x,y):
    global x_train,x_test,y_train,y_test
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

# Developing Models

**Logistic Regression**

```python
# Importing the Library
from sklearn.linear_model import LogisticRegression
```

```python
# Creating an object of the Logistic Regression
def log_reg(ob):                      # this the value for the object passed should be in strings
    global a
    ob = LogisticRegression()
    # Fitting the Data in the object
    ob.fit(x_train,y_train)
    # Predicting the values and storign it in a variable
    a = ob.predict(x_test)
#     print(a)
```

**Decision Tree**

```python
# Importing the Library
from sklearn.tree import DecisionTreeClassifier
```

```python
def dec_clf(ob):                    # this the value for the object passed should be in strings
    global b
    ob = DecisionTreeClassifier()
    # Fitting the Data in the object
    ob.fit(x_train,y_train)
    # Predicting the values and storign it in a variable
    b = ob.predict(x_test)
#     print(a)
```

**Random Forest**

```python
# Importing The Library
from sklearn.ensemble import RandomForestClassifier
```

```python
def ranf_reg(ob):                    # this the value for the object passed should be in strings
    global c
    ob = RandomForestClassifier()
    # Fitting the Data in the object
    ob.fit(x_train,y_train)
    # Predicting the values and storign it in a variable
    c = ob.predict(x_test)
#     print(a)
```

```python
# We now train test split the data.
# we call Splitting Data function which divides data in x and y
splitting_data(df)

# We divide x and y in train test split.
tts(x,y)
```

```python
# Using the Logistic Regression Model
log_reg('lr1')
```

```python
# Using Decision Tree Model
dec_clf('dt1')
```

```python
# Using Random Forest Model
ranf_reg('rf1')
```

# Model Evaluation

```python
from sklearn.metrics import classification_report
```

```python
# Classification report for Logistic Regression
print(classification_report(y_test, a))
```

```python
# Classification report for Decision Tree
print(classification_report(y_test, b))
```

```python
# Classification report for Random Forest
print(classification_report(y_test, c))
```
