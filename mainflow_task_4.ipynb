```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


sns.set(style="whitegrid")


np.random.seed(42)


n_samples = 100000
data = {
    'age': np.random.normal(loc=40, scale=10, size=n_samples).astype(int),
    'income': np.random.normal(loc=60000, scale=15000, size=n_samples),
    'expenses': np.random.normal(loc=30000, scale=10000, size=n_samples),
    'savings': np.random.normal(loc=20000, scale=5000, size=n_samples),
    'debt': np.random.normal(loc=10000, scale=3000, size=n_samples),
    'score': np.random.normal(loc=50, scale=15, size=n_samples)
}

df = pd.DataFrame(data)


print(df.describe())


df.hist(bins=30, figsize=(15, 10), layout=(3, 2))
plt.suptitle('Distribution of Variables')
plt.show()


plt.figure(figsize=(15, 10))
for i, column in enumerate(df.columns, 1):
    plt.subplot(3, 2, i)
    sns.kdeplot(df[column], shade=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()


plt.figure(figsize=(15, 10))
for i, column in enumerate(df.columns, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()


corr_matrix = df.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()


sns.pairplot(df.sample(1000))  
plt.show()


fig = px.scatter_matrix(df.sample(1000))
fig.update_traces(diagonal_visible=False)
fig.show()
```
